# 데이터 전처리 과정 1

```python
import pandas as pd


# 크롤링 한 데이터를 엑셀에서 손수 통합후 완성된 엑셀 파일 불러오기

data = pd.read_excel('all.xlsx')

df = pd.DataFrame(data)
df
```

```python
import numpy as np

# '-' 문자를 NaN으로 변환
df.replace('-', np.nan, inplace=True)
df
```

```python
# 결측치 확인 
df.isnull().sum()
```

```python
# 결측치 있는 행 제거
# 제거 이유 : 데이터가 충분히 많다고 판단
df.dropna(inplace=True)

#인덱스 다시 순서대로 하기위한 인덱스 재설정
df.reset_index(drop=True, inplace=True)

df
```

```python
# 스펙 열을 /로 분리하여 새 DataFrame 생성하는 코드
spec_df = df['스펙'].str.split('/', expand=True)

# 컬럼 수 확인 및 컬럼 이름 재정의 --> 스펙 컬럼을 (/로 분리 시켜 대학, 전공, 학점, 토익, 추가정보 등의 컬럼으로 만들기)
print("생성된 컬럼 수:", spec_df.shape[1])
spec_df.columns = ['대학', '전공', '학점', '토익', '추가 정보'] + [f'추가 정보 {i}' for i in range(5, spec_df.shape[1])]

# 기존 DataFrame에 "스펙" 열 제거 후 새로운 컬럼들 추가
df = pd.concat([df.drop('스펙', axis=1), spec_df], axis=1)
df
```

```python
# 불필요한 추가정보 5,6,7,8,9 삭제

df = df.drop(columns=['추가 정보 5'])
df = df.drop(columns=['추가 정보 6'])
df = df.drop(columns=['추가 정보 7'])
df = df.drop(columns=['추가 정보 8'])
df = df.drop(columns=['추가 정보 9'])

df
```

```python
# 불필요한 자소서, 추가정보, 토익 컬럼까지 삭제
df = df.drop(columns=['자소서'])
df = df.drop(columns=['추가 정보'])
df = df.drop(columns=['토익'])


df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix


# 입력(X)과 출력(y) 변수 분리
X = clean_df1[['학점', '전공_이과']]  # 입력 변수
y = clean_df1['대학']  # 출력 변수

# 데이터를 학습용과 테스트용으로 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# 결정트리 모델 생성 및 학습
model = DecisionTreeClassifier(random_state=42, max_depth = 4)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# 평가 척도 계산
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')
conf_matrix = confusion_matrix(y_test, y_pred)

# 결과 출력
print("정확도(Accuracy): {:.2f}".format(accuracy))
print("정밀도(Precision): {:.2f}".format(precision))
print("재현율(Recall): {:.2f}".format(recall))
print("F1 점수(F1 Score): {:.2f}".format(f1))
print("혼동 행렬(Confusion Matrix):\n", conf_matrix)
```

### 서울 TOP 10 대학 = 서연고, 서성한, 중경외시

## 통계적 기법 사용 : Z-TEST

```python
#서연고, 서성한, 중경외시 대학의 문과 비율 계산하는 코드

top_uni = clean_df3[clean_df3['대학'] == '서연고, 서성한, 중경외시']
num_humanities = (top_uni['전공'] == '문과').sum()
num_science = (top_uni['전공'] == '이과').sum()
total_students = num_humanities + num_science
humanities_ratio = num_humanities / total_students

print(f"서울 TOP 10 대학의 '문과' 학생 비율: {humanities_ratio:.2f}")



# 지방대학 문과 비율 계산하는 코드

local_uni = clean_df3[clean_df3['대학'] == '지방대']
num_humanities_local = (local_uni['전공'] == '문과').sum()
num_science_local = (local_uni['전공'] == '이과').sum()
total_students_local = num_humanities_local + num_science_local
humanities_ratio_local = num_humanities_local / total_students_local

print(f"지방대의 '문과' 학생 비율: {humanities_ratio_local:.2f}")

```

```python
# 지방대와 서울 TOP 10 대학의 문과 비율 시각화

ratios = [humanities_ratio, humanities_ratio_local]
labels = ['서연고, 서성한, 중경외시', '지방대']

# 비율 시각화
plt.figure(figsize=(8, 4))
plt.bar(labels, ratios, color=['blue', 'green'])
plt.ylabel('문과 비율')
plt.title('대학 유형별 문과 비율 비교')
plt.show()
```

```python
# 비율의 차이가 유의미한지 알수 있는 비율 차이 검정 진행
# Z -TEST 사용

from statsmodels.stats.proportion import proportions_ztest


# 문과 학생 수
count_top = top_uni['전공'].value_counts().get('문과')
count_local = local_uni['전공'].value_counts().get('문과')

# 전체 학생 수
n_top = len(top_uni)
n_local = len(local_uni)

# 비율 차이 검정
stat, pval = proportions_ztest([count_top, count_local], [n_top, n_local])

# 결과 출력
print(f"Z-statistic: {stat:.2f}, P-value: {pval}")

# 피벨류가 너무 작고 Z 스코어는 엄청 크기 때문에 귀무가설 기각
# 두 집단의 비율의 차이는 유의미한 차이가 있다 라고 검정됨 
```

**가설 3** : 지방대학의 문과 비율이 서울 TOP 10 대학의 문과 비율보다 작다 (서울 TOP 10 대학 = 서연고, 서성한, 중경외시)
        
        

**가설 설정 이유** : 문과는 특히 대학의 네임벨류가 합격에 많은 영향을 끼친다는 말이 많다. 이러한 문제가 사실인지 확인 해보기 위해.       

**가설 검증 법** : 비율 차이 검정 (Z-TEST 응용)

지방대학의 문과 비율과 서울 TOP 10 문과 비율을 가지고 시각화로 인사이트를 얻은 후 그 차이가 실제로 유의미한지 비율 차이 검정으로 확인.


**결론** : 시각화를 통해 서울 TOP 10 대학의 문과 비율이 0.46, 지방대 문과 비율은 0.19로 두배 이상 차이나는 것을 확인했다. 또한 이 차이가 유의미한지 비율 차이 검정을 통해 검증했다. 피벨류와 Z스코어를 기반으로 대립가설을 채택해 두 집단의 비율은 유의미한 차이가 있다로 결론 지었다.
      

**귀무가설** : 지방대학의 문과 비율과 서울 TOP 10 대학의 문과 비율은 유의미한 차이가 없다.
결과적으로 P-value: 4.806001276080325e-05, Z-statistic: 4.06 피벨류는 엄청 작게 나오고 Z 스코어는 엄청 크게 나와 귀무가설을 기각하고 대립가설인 "지방대학의 문과 비율과 서울 TOP 10 대학의 문과 비율은 유의미한 차이가 있다." 를 채택했다.

**최종 결론** : 시각화와, 비율 차이 검정 진행 결과 "가설 3 : 지방대학의 문과 비율이 서울 TOP 10 대학의 문과 비율보다 작다" 는 맞다고 검증되었습니다.
